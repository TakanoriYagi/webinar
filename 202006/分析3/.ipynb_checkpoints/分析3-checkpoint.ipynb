## クラスタリング

```python
%matplotlib inline
import pandas as pd
import codecs

with codecs.open("clustering.csv", "r", "Shift-JIS", "ignore") as file:
    df_es = pd.read_csv(file, delimiter=",")
    
print(df_es.head(5))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df_es2 = df_es[["ID","2_all", "3_all", "4_all", "5_all", "6_all", "7_all"]]

# 2列名以降のデータをXに、1列目のデータをyに格納
X, y = df_es2.iloc[:, 1:].values, df_es2.iloc[:, 0].values

# 平均と標準偏差を用いて標準化
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import numpy as np

# 主成分数を指定して、PCAのインスタンスを生成
pca = PCA(n_components=2)

# PCAを実行
X_pca = pca.fit_transform(X_std)

print(np.sum(pca.explained_variance_ratio_))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=1000,
                tol=1e-04,
                random_state=0)
    km.fit(X_pca)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
km = KMeans(n_clusters=3,
           init='k-means++',  # セントロイドの初期値をランダムに選択
           n_init=10,      # 異なるセントロイドの初期値をランダムに選択
           max_iter=1000,   # k-meansアルゴリズム内部の最大イテレーション回数
           tol=1e-04,
           random_state=0)
y_km = km.fit_predict(X_pca)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_pca[y_km==0,0],       # グラフのxの値
           X_pca[y_km==0,1],        # グラフのyの値
           s=50,                # プロットサイズ
           c='lightgreen',      # プロットの色
           edgecolor='black',   # プロットの線の色
           marker='s',
           label='cluster 1')

plt.scatter(X_pca[y_km==1,0],
           X_pca[y_km==1,1],
           s=50,
           c='orange',
           edgecolor='black',
           marker='o',
           label='cluster 2')

plt.scatter(X_pca[y_km==2,0],
           X_pca[y_km==2,1],
           s=50,
           c='lightblue',
           edgecolor='black',
           marker='v',
           label='cluster 3')

plt.scatter(km.cluster_centers_[:,0],
           km.cluster_centers_[:,1],
           s=250,
           c='red',
           edgecolor='black',
           marker='*',
           label='centroids')

plt.legend(scatterpoints=1)
plt.grid()
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.tight_layout()
plt.show()
```

## 重要度分析

```python
with codecs.open("importance.csv", "r", "Shift-JIS", "ignore") as file:
    df_es3 = pd.read_csv(file, delimiter=",")
    
print(df_es3.head(5))
```

```python
all = {}
all["id"] = df_es3["ID"]
all["data"] = df_es3.iloc[:,1:6].values
all["target"] = df_es3.iloc[:,6].values
all["feature_names"] = df_es3.iloc[:,1:6].columns
```

```python
all["feature_names"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all["data"], all["target"])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rnd_reg = RandomForestRegressor(n_estimators=500, n_jobs=-1)
rnd_reg.fit(X_train, y_train)
ls = []
for name, score in zip(all["feature_names"], rnd_reg.feature_importances_):
    ls.append([name,score])

ls_sort = sorted(ls, reverse=True, key=lambda x: x[1])

print("重要度")

for i in range(5):
    print(ls_sort[i][0], ls_sort[i][1])

y_pred_rf = rnd_reg.predict(X_test)
r2_score = r2_score(y_test, y_pred_rf)

print("\nr2_score:",r2_score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

boost_reg = GradientBoostingRegressor(loss="huber",learning_rate=0.01, n_estimators=500, max_features=2)
boost_reg.fit(X_train, y_train)
ls = []
for name, score in zip(all["feature_names"], boost_reg.feature_importances_):
    ls.append([name,score])

ls_sort = sorted(ls, reverse=True, key=lambda x: x[1])

print("重要度")

for i in range(5):
    print(ls_sort[i][0], ls_sort[i][1])

y_pred_rf = boost_reg.predict(X_test)
r2_score = r2_score(y_test, y_pred_rf)

print("\nr2_score:",r2_score)
```
